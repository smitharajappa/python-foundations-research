# Data Structure Lookup Experiment

This notebook explores performance differences between Python lists and dictionaries.


```
import time

data_list = list(range(1000000))
data_dict = {i: i for i in range(1000000)}

start = time.time()
999999 in data_list
list_time = time.time() - start

start = time.time()
999999 in data_dict
dict_time = time.time() - start

list_time, dict_time
```

## Observation
Dictionary lookups are significantly faster than list lookups for large datasets.
